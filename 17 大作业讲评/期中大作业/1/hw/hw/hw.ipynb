```python
# ç­›é€‰å‡ºè§‚å½±äººæ¬¡å¤§äº300çš„ç”µå½±
popular = ratings['movie_id'].value_counts()
popular = popular[popular > 300]
popular = popular.rename('count')
popular = popular.rename_axis('movie_id')
print("Popular:")
display(popular)
```

```python
import matplotlib.pyplot as plt

# å¦‚ä½•è¡¡é‡åå¥½ç¨‹åº¦ï¼Ÿçœ‹è¿‡æŸéƒ¨ç”µå½±çš„ç”·æ€§/å¥³æ€§è§‚ä¼—å¯¹è¯¥éƒ¨ç”µå½±çš„å¹³å‡æ‰“åˆ†æ˜¾ç„¶æ˜¯æœ€é‡è¦çš„è¡¡é‡ä¾æ®ï¼Œ
# ä½†æˆ‘ä»¬éœ€è¦æ³¨æ„åˆ°ï¼Œå—â€œåå¥½â€ç¨‹åº¦ç”¨ç”·æ€§/å¥³æ€§è§‚ä¼—å¯¹è¯¥éƒ¨ç”µå½±çš„å¹³å‡æ‰“åˆ†ä¹‹å·®æ¥è¡¡é‡è¦æ¯”ç”¨å„è‡ªçš„ç»å¯¹å€¼æ¥è¡¡é‡æ›´åˆç†ã€‚
# ä½†æ˜¯åŒæ—¶ï¼Œè¯¥ç”µå½±çš„è§‚å½±äººæ¬¡ä¹Ÿåº”å½“æœ‰åŠ æƒï¼Œæ¯”å¦‚æ€»è§‚å½±äººæ¬¡éå¸¸å¤šçš„ç”µå½±å¦‚æœè¿˜å‘ˆç°å‡ºæ˜æ˜¾çš„æ€§åˆ«æ‰“åˆ†å·®å¼‚ï¼Œ
# é‚£ä¹ˆæ›´èƒ½è¯´æ˜è¿™éƒ¨ç”µå½±æ˜¯æ˜æ˜¾çš„æ€§åˆ«åå¥½ç”µå½±ï¼Œå› ä¸ºæ‰“åˆ†éšæœºæ€§ç­‰å¯èƒ½å¯¼è‡´éå†…åœ¨çš„ç”·å¥³æ‰“åˆ†å·®å¼‚çš„å› ç´ å½±å“ä¼šè¾ƒå¼±ï¼Œæ€§åˆ«å·®å¼‚ç»“æœçš„â€œä¿¡åº¦â€å°±æ›´é«˜ã€‚
# ä¸è¿‡è¿™ä¸€é¡¹åªæ˜¯è¾…åŠ©æŒ‡æ ‡ï¼Œä¸åº”æˆä¸ºå½±å“æœ€ç»ˆè¯„ä»·å‡½æ•°çš„æœ€ä¸»è¦å› ç´ ã€‚æ•…æœ€ç»ˆæˆ‘é€‰æ‹©ç”¨ " * ln(æ€»è§‚å½±äººæ¬¡) " çš„æ–¹å¼ã€‚
# å› æ­¤ï¼Œæˆ‘æœ€ç»ˆå®šä¹‰çš„è¯„ä»·å‡½æ•°æ˜¯ " (ç”·æ€§è§‚ä¼—å¹³å‡æ‰“åˆ† - å¥³æ€§è§‚ä¼—å¹³å‡æ‰“åˆ†) * ln(æ€»è§‚å½±äººæ¬¡) " ã€‚

# æˆ‘è¿™é‡Œæ²¡æœ‰ä½¿ç”¨æ¯éƒ¨ç”µå½±çš„â€œç”·æ€§è§‚å½±äººæ¬¡â€å’Œâ€œå¥³æ€§è§‚å½±äººæ¬¡â€è¿™ä¸¤ä¸ªæŒ‡æ ‡ï¼Œ
# æ˜¯å› ä¸ºè¿™ä¼šæ¶‰åŠç”·/å¥³æ€§è§‚ä¼—äººæ•°ä¸ä¸€è‡´ç­‰é—®é¢˜ï¼Œä¼šå¸¦æ¥éº»çƒ¦ï¼Œçºµç„¶è¿›è¡Œå½’ä¸€åŒ–ç­‰æ“ä½œä¹Ÿå¾ˆéš¾ä¸ä½¿è¯„ä»·ç»“æœå¤±çœŸã€‚

gender_users = pd.merge(users, ratings, on='user_id', how='outer')
gender_users = gender_users[gender_users['movie_id'].isin(popular.index)] # åªä¿ç•™popularçš„ç”µå½±

female_users = gender_users[gender_users['gender'] == 'F']
male_users = gender_users[gender_users['gender'] == 'M']

# è®¡ç®—å¹³å‡æ‰“åˆ†
female_mean_rating = female_users.groupby('movie_id')['rating'].mean()
male_mean_rating = male_users.groupby('movie_id')['rating'].mean()
print("Female mean rating:")
display(female_mean_rating)
print("\nMale mean rating:")
display(male_mean_rating)

# è®¡ç®—è¯„ä»·å‡½æ•°
score = male_mean_rating - female_mean_rating
score = pd.merge(score, popular, on='movie_id', how='outer') # åŠ å…¥æ€»è§‚å½±äººæ¬¡è¿™ä¸€åˆ—
score['score'] = score['rating'] * np.log(score['count']) # score['score']è¿™ä¸€åˆ—ä¸ºè¯„ä»·å‡½æ•°
print("\nScore:")
display(score)
```

```python
# ç­›å‡ºå‰20åå’Œå20åçš„ç”µå½±ï¼Œåˆ†åˆ«ä½œä¸ºç”·/å¥³æ€§åå¥½çš„ç”µå½±
male_preference = score.nlargest(20, 'score').reset_index()
female_preference = score.nsmallest(20, 'score').reset_index()
female_preference['score'] = - female_preference['score']
```

```python
# è¯¦ç»†å±•ç¤ºè¿™äº›é€‰å‡ºçš„ç”µå½±çš„ä¿¡æ¯
print("ç”·æ€§æœ€åå¥½çš„20éƒ¨ç”µå½±ï¼š")
male_preference_detail = pd.merge(male_preference, movies, on='movie_id', how='inner')
male_preference_detail = male_preference_detail.drop(['rating', 'count', 'score', 'movie_id'], axis=1) # å»é™¤æ— éœ€å±•ç¤ºçš„åˆ—
male_preference_detail['ranking'] = male_preference_detail.index.to_series().apply(lambda x: x+1)
male_preference_detail = male_preference_detail.set_index('ranking') # å¢åŠ â€œæ’åâ€è¿™ä¸€åˆ—ï¼Œå¹¶å°†å…¶è®¾ä¸ºindex
display(male_preference_detail)

print("\nå¥³æ€§æœ€åå¥½çš„20éƒ¨ç”µå½±ï¼š")
female_preference_detail = pd.merge(female_preference, movies, on='movie_id', how='inner')
female_preference_detail = female_preference_detail.drop(['rating', 'count', 'score', 'movie_id'], axis=1) # å»é™¤æ— éœ€å±•ç¤ºçš„åˆ—
female_preference_detail['ranking'] = female_preference_detail.index.to_series().apply(lambda x: x+1)
female_preference_detail = female_preference_detail.set_index('ranking') # å¢åŠ â€œæ’åâ€è¿™ä¸€åˆ—ï¼Œå¹¶å°†å…¶è®¾ä¸ºindex
display(female_preference_detail)
```

```python
# å…¶ä»–çš„æµè¡Œç»Ÿè®¡é‡
# ğ‘…ğ‘=ğ‘Šğ‘…+(1âˆ’ğ‘Š)ğ‘…0
# ğ‘…ğ‘=å¹³å‡(è´å¶æ–¯)è¯„çº§
# ğ‘…=ä¸ªä½“è¯„åˆ†ï¼šæ­¤ä¸ªä½“çš„å¹³å‡è¯„åˆ†ã€‚
# ğ‘…0=å…ˆéªŒè¯„çº§ï¼šå…¨å±€å¹³å‡è¯„çº§ï¼Œé€‚ç”¨äºæ•°æ®åº“ä¸­çš„æ‰€æœ‰é¡¹ç›®ã€‚
# ğ‘Š=åŠ æƒå› å­ï¼šå¦‚æœæ­¤é¡¹æŠ•ç¥¨è¾ƒå°‘ï¼Œåˆ™åº”è¶‹äº0ï¼Œå¦‚æœæŠ•ç¥¨è¾ƒå¤šï¼Œåˆ™åº”è¶‹äº1ã€‚
# ğ‘Š=ğ‘šğ‘ğ‘¥(ğ›¼*ğ‘›/ğ‘ğ‘ğ‘£,1)ï¼Œnä¸ºè¯¥ç”µå½±çš„è§‚å½±äººæ¬¡ï¼Œğ‘ğ‘ğ‘£ä¸ºæ‰€æœ‰ç”µå½±çš„å¹³å‡è§‚å½±äººæ¬¡ï¼Œğ›¼ä¸º0.5åˆ°1ä¹‹é—´çš„æŸä¸ªæ•°

#é‡æ–°è®¡ç®—popular
R=ratings.groupby('movie_id')['rating'].mean()
N=ratings.groupby('movie_id')['rating'].count()
popular=pd.DataFrame({'rating':R,'count':N})
W=popular['count'].apply(lambda x: max(0.5*x/popular['count'].mean(),1))
popular['rating']=W*popular['rating']+(1-W)*popular['rating'].mean()
popular=popular.sort_values(by='rating',ascending=False)
popular_percentile=popular['rating'].quantile(0.9)
popular=popular[popular['rating']>=popular_percentile]
print(popular)

#å‰©ä¸‹çš„æ­¥éª¤é¢˜ç›®æ²¡è¦æ±‚ï¼Œå°±å’Œä¸Šé¢ç”¨ä¸€æ ·çš„ï¼Œè™½ç„¶ä¹Ÿå¯ä»¥ä¼˜åŒ–æŒºå¤šï¼Œä½†æ˜¯ä¹Ÿè¶³å¤Ÿäº†
gender_users = pd.merge(users, ratings, on='user_id', how='outer')
gender_users = gender_users[gender_users['movie_id'].isin(popular.index)] # åªä¿ç•™popularçš„ç”µå½±

female_users = gender_users[gender_users['gender'] == 'F']
male_users = gender_users[gender_users['gender'] == 'M']

# è®¡ç®—å¹³å‡æ‰“åˆ†
female_mean_rating = female_users.groupby('movie_id')['rating'].mean()
male_mean_rating = male_users.groupby('movie_id')['rating'].mean()
print("Female mean rating:")
display(female_mean_rating)
print("\nMale mean rating:")
display(male_mean_rating)

# è®¡ç®—è¯„ä»·å‡½æ•°
score = male_mean_rating - female_mean_rating
score = pd.merge(score, popular, on='movie_id', how='outer') # åŠ å…¥æ€»è§‚å½±äººæ¬¡è¿™ä¸€åˆ—
score['score'] = score['rating_x'] * score['rating_y'] # score['score']è¿™ä¸€åˆ—ä¸ºè¯„ä»·å‡½æ•°
print("\nScore:")
display(score)

# ç­›å‡ºå‰20åå’Œå20åçš„ç”µå½±ï¼Œåˆ†åˆ«ä½œä¸ºç”·/å¥³æ€§åå¥½çš„ç”µå½±
male_preference = score.nlargest(20, 'score').reset_index()
female_preference = score.nsmallest(20, 'score').reset_index()
female_preference['score'] = - female_preference['score']

# è¯¦ç»†å±•ç¤ºè¿™äº›é€‰å‡ºçš„ç”µå½±çš„ä¿¡æ¯
print("ç”·æ€§æœ€åå¥½çš„20éƒ¨ç”µå½±ï¼š")
male_preference_detail = pd.merge(male_preference, movies, on='movie_id', how='inner')
male_preference_detail = male_preference_detail.drop(['rating_x', 'rating_y', 'score', 'movie_id'], axis=1) # å»é™¤æ— éœ€å±•ç¤ºçš„åˆ—
male_preference_detail['ranking'] = male_preference_detail.index.to_series().apply(lambda x: x+1)
male_preference_detail = male_preference_detail.set_index('ranking') # å¢åŠ â€œæ’åâ€è¿™ä¸€åˆ—ï¼Œå¹¶å°†å…¶è®¾ä¸ºindex
display(male_preference_detail)

print("\nå¥³æ€§æœ€åå¥½çš„20éƒ¨ç”µå½±ï¼š")
female_preference_detail = pd.merge(female_preference, movies, on='movie_id', how='inner')
female_preference_detail = female_preference_detail.drop(['rating_x', 'rating_y', 'score', 'movie_id'], axis=1) # å»é™¤æ— éœ€å±•ç¤ºçš„åˆ—
female_preference_detail['ranking'] = female_preference_detail.index.to_series().apply(lambda x: x+1)
female_preference_detail = female_preference_detail.set_index('ranking') # å¢åŠ â€œæ’åâ€è¿™ä¸€åˆ—ï¼Œå¹¶å°†å…¶è®¾ä¸ºindex
display(female_preference_detail)
```

```python
import os
import cv2
import numpy as np
import pandas as pd
from PIL import Image
import torch
from img2vec_pytorch import Img2Vec



# æ–‡ä»¶å¤¹è·¯å¾„
folder_path = "./poster"

# åˆå§‹åŒ–ç‰¹å¾åˆ—è¡¨
features = []

img2vec_model = Img2Vec()
# éå†æ–‡ä»¶å¤¹ä¸­çš„æ¯ä¸ªæµ·æŠ¥å›¾åƒ
for filename in os.listdir(folder_path):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # è¯»å–å›¾åƒ
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)

        # æå–é¢œè‰²ç›´æ–¹å›¾å’Œç°åº¦ç›´æ–¹å›¾ç‰¹å¾
        image = Image.open(image_path)
        if image.mode != 'RGB':
            image = image.convert('RGB')
        vector = img2vec_model.get_vec(image)

        feature_vector = vector

        # æ·»åŠ å›¾åƒæ–‡ä»¶å‰ç¼€ï¼ˆç”µå½±IDï¼‰å’Œç‰¹å¾å‘é‡åˆ°ç‰¹å¾åˆ—è¡¨ä¸­
        features.append([filename.split('.')[0], feature_vector])

# å°†ç‰¹å¾åˆ—è¡¨è½¬æ¢ä¸ºDataFrame
columns = ['movie_id', 'features']
df = pd.DataFrame(features, columns=columns)

# è¾“å‡ºå¸¦æœ‰ç”µå½±IDå’Œç‰¹å¾å‘é‡çš„DataFrame
print("DataFrame with Movie ID and Features:")
print(df.head())
```

```python
import pandas as pd

# è¯»å–CSVæ–‡ä»¶
movies_df = pd.read_csv("./movies.csv")

# æå–genreä¿¡æ¯
genres = movies_df['genres']
# åˆå§‹åŒ–genreçš„å­—å…¸
genre_dict = {}

# éå†æ¯ä¸ªç”µå½±çš„genreä¿¡æ¯
for genre_string in genres:
    # å°†genreå­—ç¬¦ä¸²æŒ‰ç…§"|"åˆ†å‰²æˆåˆ—è¡¨
    genre_list = genre_string.split("|")
    # éå†åˆ—è¡¨ä¸­çš„æ¯ä¸ªgenre
    for genre in genre_list:
        # å°†genreæ·»åŠ åˆ°å­—å…¸ä¸­ï¼Œå¦‚æœå·²å­˜åœ¨åˆ™ä¸é‡å¤æ·»åŠ 
        if genre not in genre_dict:
            genre_dict[genre] = len(genre_dict)

# è¾“å‡ºgenreå­—å…¸
print("Genre Dictionary:")
print(genre_dict)
```
