2.1补全下列常用的激活函数及其导数,熟记并画出它们的图像（2分），你也可以查阅资料后补充一些其它常用的激活函数（不计分）

```python
import numpy as np
```

```python
def sigmoid(x):
    #todo
    return 1.0/(1.0 + np.exp(-x))

def d_sigmoid(x):
    #todo
    return sigmoid(x)*(1-sigmoid(x))
    
def ReLU(x):
    #todo
    return np.maximum(0, x)

def d_ReLU(x):
    #todo
    return np.where(x>0, 1, 0)

def Tanh(x):
    #todo
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

def d_Tanh(x):
    #todo
    return 1-Tanh(x)**2
```

```python
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y1 = sigmoid(x)
y2 = d_sigmoid(x)
y3 = ReLU(x)
y4 = d_ReLU(x)
y5 = Tanh(x)
y6 = d_Tanh(x)

plt.figure(figsize=(6,9))

plt.subplot(321)
plt.plot(x, y1)
plt.title('Sigmoid')

plt.subplot(322)
plt.plot(x, y2)
plt.title('d_Sigmoid')

plt.subplot(323)
plt.plot(x, y3)
plt.title('ReLU')

plt.subplot(324)
plt.plot(x, y4)
plt.title('d_ReLU ')

plt.subplot(325)
plt.plot(x, y5)
plt.title('Tanh')

plt.subplot(326)
plt.plot(x, y6)
plt.title('d_Tanh')

plt.tight_layout()
plt.show()
```
